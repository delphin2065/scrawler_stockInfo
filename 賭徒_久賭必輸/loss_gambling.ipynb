```python
import numpy as np
from matplotlib import pyplot as plt
```

```python
fig = plt.figure(figsize=(10, 5))
ax = fig.add_subplot(1, 1, 1)

cnt = 0
for i in range(10):
  # 設定賭局參數
  initial_money = 1000  
  bet_amount = 10
  win_probability = 0.5
  target_profit = 5000
  target_loss = 0  
  profit_loss_ratio = 1

  # 模擬賭局
  money = initial_money
  money_history = [money]
  bet_count = 0

  while money > target_loss and money < initial_money + target_profit:
      bet_result = np.random.choice([1, -1], p=[win_probability, 1 - win_probability])
      
      if bet_result == 1:
        money = money + bet_amount * profit_loss_ratio * bet_result

      if bet_result == -1:
        money = money + bet_amount * bet_result


      money_history.append(money)
      bet_count += 1


  cnt = cnt + 1
  ax.plot(money_history, label='第' + str(cnt) + '次模擬')
  print(f"第{str(cnt)}次模擬': 賭博次數: {bet_count}, 最終資金: {money}")
  print('='*50)
  
ax.set_xlabel('賭博次數')
ax.set_xlabel('剩餘資金')
ax.set_title('賭博結果模擬')
ax.axhline(initial_money, color='r', linestyle='--', label='初始資金')
ax.axhline(initial_money + target_profit, color='g', linestyle='--', label='目標資金')
plt.legend()
plt.show()
```
